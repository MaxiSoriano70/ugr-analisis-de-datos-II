# Introducción a ggplot2

`ggplot2` Es una libraría que pertenece al paquete `Tidyverse`. Para más información visitar:
https://www.tidyverse.org/

Es amplamiente utilizada para realizar gráficos, particularmente en ciencia de datos. En el siguiente link encontrarán información detallada sobre esta librería:
https://www.rdocumentation.org/packages/ggplot2/versions/3.4.3


### Componentes

La gramática de los gráficos se explica mediante los siguientes componentes.


`Data` Es el dataset que vamos a utilizar

`geoms` Son los puntos, líneas, cajas entre otros elementos que se agregarán a los gráficos

`aes` Personalizar la visual de los gráficos

`Stats` Son las transformaciones estadísticas y nos ayudan a comprender y visualizar la información

`Scales` Refieren al coloreo y/o escalado de los datos. Refieren a los ejes y leyendas

`Facets` Se utilizan para particionar los datasets y subdividir los gráficos

`Themes` Permiten controlar la apariencia general de todos los elementos que no son datos (ejemplo, los ejes y el color de fondo)


### Primeros pasos


```r
library(ggplot2)
```

```r
#Inicio del ejemplo: Dataset Flores Iris----
data("iris")
attach(iris)
iris
dim(iris)
help(iris)
head(iris)
tail(iris)
View(iris)
```

```r
summary(iris)
#Flores iris con ggplot2----
#Este gráfico contiene los ejes que especificamos pero no contiene los datos
ggplot(iris,
       aes(Sepal.Length, Petal.Length))
```

```r
#Agregamos los datos
ggplot(iris,
       aes(Sepal.Length, Petal.Length)) + geom_point()

#Agregamos línea de tendencia
ggplot(iris,
       aes(Sepal.Length, Petal.Length)) + geom_point() +
  geom_smooth(method = "lm")

#Agregamos color por especie:
ggplot(iris, aes(Sepal.Length, Petal.Length, color = Species)) +
  geom_point() +
  geom_smooth(method = "lm")
```

### aes


```r

#aes----
ggplot(iris, aes(Sepal.Length, Petal.Length)) +
  geom_point(aes(color = Species))

#Agregamos línea de tendencia. Notar la diferencia según posición del aes()
ggplot(iris, aes(Sepal.Length, Petal.Length)) +
  geom_point(aes(color = Species))+
  geom_smooth(method="lm")

```

### geom


```r
#geom----
# Largo de los sépalos por especie

#Gráfico de puntos
p1 <- ggplot(iris, aes(Species, Sepal.Length)) +
  geom_point()
p1

#Gráfico de puntos con ruido en el eje horizontal
p2 <- ggplot(iris, aes(Species, Sepal.Length)) +
  geom_jitter(width = 0.1)
p2

# Boxplot
bp1 <- ggplot(iris, aes(Species, Sepal.Length)) +
  geom_boxplot()
bp1

# Violin
v1 <- ggplot(iris, aes(Species, Sepal.Length)) +
  geom_violin()
v1

# Combinando boxplot y violin

comb1 <- ggplot(iris, aes(Species, Sepal.Length)) +
  geom_violin(fill='blue', alpha=0.45) +
  geom_boxplot(color="lightblue", fill="white",
               lwd=0.6, width=0.15)
comb1

# Grafico de barras con medias
b1 <- ggplot(iris, aes(Species, Sepal.Length)) +
  stat_summary(fun.y = mean, geom = "bar",
               width=0.2) +
  stat_summary(fun.data = mean_se, geom = "errorbar",
               color="red", width=0.2)
b1
```

### facet


```r
#Facet----
ggplot(iris, aes(Sepal.Length, Petal.Length)) +
  geom_point()+
  facet_wrap(~Species)


ggplot(iris, aes(Sepal.Length, Petal.Length)) +
  geom_point(aes(color=Petal.Width))+
  facet_wrap(~Species)
```

### cowplot


```r

#cowplot----
#Si es la primera vez

install.packages("cowplot")
#Graficar dos gráficos en el mismo panel

cowplot::plot_grid(p1,p2,bp1,v1,comb1,b1, labels = "AUTO")


```
